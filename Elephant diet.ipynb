Οι φύλακες ενός ζωολογικού κήπου θέλουν να εφαρμόσουν ένα πρόγραμμα διατροφής σε έναν λαίμαργο ελέφαντα, έτσι ώστε αφενός να διατηρηθεί το βάρος του σε ένα επιθυμητό επίπεδο και αφετέρου να λαμβάνει τη μέγιστη δυνατή ποσότητα τροφής για τη σωστή ανάπτυξή του. Για το λόγο αυτό, οι διαιτολόγοι έχουν δώσει ένα πρόγραμμα διατροφής με την καθημερινή ποσότητα τροφής που πρέπει να λαμβάνει ο ελέφαντας, αλλά με τον περιορισμό ότι δεν μπορεί να λαμβάνει τροφή για τρεις συνεχόμενες ημέρες. Το πρόβλημα των φυλάκων είναι να βρουν τις ημέρες εκείνες που πρέπει να λάβει τροφή ο ελέφαντας μεγιστοποιώντας την ποσότητα των κιλών τροφής, υπό τον συγκεκριμένο περιορισμό (της μη λήψης τροφής για τρεις συνεχόμενες ημέρες). Το πρόγραμμα διατροφής n ημερών
δίνεται ως ένα διάνυσμα Α, στο οποίο η θέση
A(i) περιέχει τα κιλά τροφής που πρέπει να λάβει ο ελέφαντας την i-οστή ημέρα, 1 ≤ i ≤ n.
Στο κάτωθι πρόγραμμα πατήστε Shift-Enter βάλτε πρώτα τις μέρες και μετά τα κιλά διατροφής κάθε μερας και θα υπολογισθεί αυτόματα το πιό αποδοτικό σενάριο.

```python
A=[]                    # Αρχικοποιούμε την λίστα για την εισαγωγή των ακεραίων που θα εισάγουμε                      
N=int(input('Εισάγετε το πλήθος των ημερών διατροφής του ελέφαντα N= '))  # Εισαγωγή του πλήθους που επιθυμούμε

for i in range(N):                                                         # Εισαγωγή ενός - ενός των ακεραίων
    x=(input('Δώστε τα κιλά διατροφής για την '+str(i+1)+'η ημέρα '));
    A.append(int(x))
print("Πίνακας με τα κιλά ανά μέρα ", A)
n = len(A)
B = [0 for i in range(n)]
B[0] = A[0] 
B[1] = A[0] + A[1] 
B[2] = max(A[0]+A[1],A[0]+A[2],A[1]+A[2])
for i in range(3, n): 
    B[i] = max(B[i - 1], A[i] + B[i - 2] , A[i] + B[i - 1] - A[i - 2], A[i] +A[i-1] + B[i - 3])

for i in range(2,n): 
    print("Οι καλύτερες τιμές για τον πίνακα:  " + str(A[:i+1]) + " είναι " + 'B[' + str(i) + '] = ' + str(B[i]))
```
